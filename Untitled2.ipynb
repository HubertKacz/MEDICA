```python
from flask import Flask, render_template_string, request, redirect, session, url_for
import sqlite3
from google.colab.output import serve_kernel_port_as_window

app = Flask(__name__)
app.secret_key = 'medica_secure_key_2025' # Klucz dla sesji i bezpieczeństwa danych

# --- 1. BAZA WBUDOWANA (Embedded Database) ---
# Rozwiązuje problem rozproszonych baz w placówkach [cite: 19, 22]
def init_db():
    conn = sqlite3.connect('medica_central.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS pacjenci (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            imie TEXT NOT NULL,
            nazwisko TEXT NOT NULL,
            pesel TEXT UNIQUE NOT NULL,
            haslo TEXT NOT NULL,
            historia_medyczna TEXT
        )
    ''')
    conn.commit()
    conn.close()

init_db()

# --- 2. INTERFEJS HTML ---
# Projekt intuicyjny, aby nie wymagał wielogodzinnych szkoleń [cite: 31, 33]
base_layout = """
<!DOCTYPE html>
<html lang="pl">
<head>
    <meta charset="UTF-8">
    <title>Medica+ Centralny System Obsługi</title>
    <style>
        body { font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; background-color: #eef2f3; color: #333; }
        .container { max-width: 500px; margin: 50px auto; background: white; padding: 30px; border-radius: 12px; box-shadow: 0 4px 20px rgba(0,0,0,0.08); }
        h2 { color: #005b96; text-align: center; }
        input, textarea { width: 100%; padding: 12px; margin: 8px 0; border: 1px solid #ccc; border-radius: 6px; box-sizing: border-box; }
        button { background-color: #005b96; color: white; padding: 12px; border: none; border-radius: 6px; cursor: pointer; width: 100%; font-size: 16px; }
        button:hover { background-color: #004a7c; }
        .nav { text-align: center; margin-bottom: 20px; }
        .nav a { margin: 0 15px; text-decoration: none; color: #005b96; font-weight: bold; }
        .error { color: #d9534f; text-align: center; font-weight: bold; }
    </style>
</head>
<body>
    <div class="nav">
        <a href="/">Strona Główna</a> | <a href="/rejestracja">Nowy Pacjent</a> | <a href="/logowanie">Logowanie</a>
    </div>
    <div class="container">
        {% block content %}{% endblock %}
    </div>
</body>
</html>
"""

# --- 3. LOGIKA SYSTEMU ---

@app.route('/')
def home():
    return render_template_string(base_layout + "<h2>Witaj w Medica+</h2><p>Zintegrowany system obsługi pacjentów dla wszystkich 5 placówek.</p>")

# Formularz dodawania pacjenta (Realizacja punktu 2 zadania)
@app.route('/rejestracja', methods=['GET', 'POST'])
def rejestracja():
    if request.method == 'POST':
        try:
            conn = sqlite3.connect('medica_central.db')
            cursor = conn.cursor()
            cursor.execute('INSERT INTO pacjenci (imie, nazwisko, pesel, haslo, historia_medyczna) VALUES (?, ?, ?, ?, ?)',
                           (request.form['imie'], request.form['nazwisko'], request.form['pesel'], request.form['haslo'], request.form['historia']))
            conn.commit()
            conn.close()
            return redirect(url_for('logowanie'))
        except sqlite3.IntegrityError:
            return render_template_string(base_layout + '<p class="error">PESEL już istnieje w systemie!</p>')

    html = """
    {% extends "base" %}
    {% block content %}
    <h2>Rejestracja Pacjenta</h2>
    <form method="POST">
        <input type="text" name="imie" placeholder="Imię" required>
        <input type="text" name="nazwisko" placeholder="Nazwisko" required>
        <input type="text" name="pesel" placeholder="PESEL" required>
        <input type="password" name="haslo" placeholder="Hasło" required>
        <textarea name="historia" placeholder="Historia choroby (dane medyczne)"></textarea>
        <button type="submit">Zapisz do bazy centralnej</button>
    </form>
    {% endblock %}
    """
    return render_template_string(base_layout + html)

# Mechanizm logowania (Realizacja punktu 3 zadania)
@app.route('/logowanie', methods=['GET', 'POST'])
def logowanie():
    error = None
    if request.method == 'POST':
        conn = sqlite3.connect('medica_central.db')
        cursor = conn.cursor()
        cursor.execute('SELECT id FROM pacjenci WHERE pesel = ? AND haslo = ?', (request.form['pesel'], request.form['haslo']))
        user = cursor.fetchone()
        conn.close()
        if user:
            session['user_id'] = user[0]
            return redirect(url_for('karta_pacjenta'))
        error = "Niepoprawne dane logowania."

    html = """
    {% extends "base" %}
    {% block content %}
    <h2>Logowanie Pacjenta</h2>
    {% if error %}<p class="error">{{ error }}</p>{% endif %}
    <form method="POST">
        <input type="text" name="pesel" placeholder="PESEL" required>
        <input type="password" name="haslo" placeholder="Hasło" required>
        <button type="submit">Zaloguj się</button>
    </form>
    {% endblock %}
    """
    return render_template_string(base_layout + html, error=error)

# Karta pacjenta (Realizacja punktu 4 zadania)
@app.route('/karta')
def karta_pacjenta():
    if 'user_id' not in session: return redirect(url_for('logowanie'))

    conn = sqlite3.connect('medica_central.db')
    cursor = conn.cursor()
    cursor.execute('SELECT imie, nazwisko, pesel, historia_medyczna FROM pacjenci WHERE id = ?', (session['user_id'],))
    p = cursor.fetchone()
    conn.close()

    html = """
    {% extends "base" %}
    {% block content %}
    <h2>Karta Pacjenta</h2>
    <p><strong>Imię i Nazwisko:</strong> {{ p[0] }} {{ p[1] }}</p>
    <p><strong>Numer PESEL:</strong> {{ p[2] }}</p>
    <div style="background: #f9f9f9; padding: 15px; border-left: 5px solid #005b96; margin-top: 20px;">
        <h3>Historia Medyczna</h3>
        <p>{{ p[3] if p[3] else "Brak historii." }}</p>
    </div>
    <br><a href="/logout"><button style="background-color:#777;">Wyloguj</button></a>
    {% endblock %}
    """
    return render_template_string(base_layout + html, p=p)

@app.route('/logout')
def logout():
    session.pop('user_id', None)
    return redirect(url_for('home'))

# --- 4. URUCHOMIENIE W GOOGLE COLAB ---
serve_kernel_port_as_window(5000)
app.run(port=5000)
```

```python
from flask import Flask, render_template_string, request, redirect, session, url_for
import sqlite3
from google.colab.output import serve_kernel_port_as_window

app = Flask(__name__)
app.secret_key = 'medica_plus_2026_secure'

# --- 1. INICJALIZACJA BAZY WBUDOWANEJ ---
def init_db():
    # Tworzymy centralną bazę danych, aby rozwiązać problem braku przepływu informacji między placówkami [cite: 19, 22]
    conn = sqlite3.connect('medica_plus.db')
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS pacjenci (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            imie TEXT NOT NULL,
            nazwisko TEXT NOT NULL,
            pesel TEXT UNIQUE NOT NULL,
            haslo TEXT NOT NULL,
            historia_medyczna TEXT
        )
    ''')
    conn.commit()
    conn.close()

init_db()

# --- 2. LAYOUT I SZABLONY HTML ---
# Projekt zgodny z potrzebą intuicyjności dla personelu i lekarzy [cite: 31, 33]
LAYOUT = """
<!DOCTYPE html>
<html>
<head>
    <title>Medica+ System Centralny</title>
    <style>
        body { font-family: Arial, sans-serif; background-color: #f4f7f6; display: flex; flex-direction: column; align-items: center; }
        .nav { background: #005b96; width: 100%; padding: 15px; text-align: center; margin-bottom: 30px; }
        .nav a { color: white; margin: 0 15px; text-decoration: none; font-weight: bold; }
        .box { background: white; padding: 25px; border-radius: 8px; box-shadow: 0 4px 10px rgba(0,0,0,0.1); width: 400px; }
        h2 { color: #005b96; margin-top: 0; }
        input, textarea { width: 100%; padding: 10px; margin: 10px 0; border: 1px solid #ddd; border-radius: 4px; box-sizing: border-box; }
        button { background-color: #005b96; color: white; padding: 10px; border: none; border-radius: 4px; cursor: pointer; width: 100%; font-size: 16px; }
        .error { color: red; font-size: 14px; }
    </style>
</head>
<body>
    <div class="nav">
        <a href="/">Start</a> | <a href="/rejestracja">Nowy Pacjent</a> | <a href="/logowanie">Logowanie</a>
    </div>
    <div class="box">
        {{ content | safe }}
    </div>
</body>
</html>
"""

# --- 3. LOGIKA APLIKACJI ---

@app.route('/')
def index():
    html = "<h2>Witaj w Medica+</h2><p>Zintegrowany system dla wszystkich 5 placówek warszawskich.</p>"
    return render_template_string(LAYOUT, content=html)

@app.route('/rejestracja', methods=['GET', 'POST'])
def rejestracja():
    error = ""
    if request.method == 'POST':
        # Rejestracja pacjenta online - rozładowanie kolejek telefonicznych [cite: 14, 18]
        imie = request.form['imie']
        nazwisko = request.form['nazwisko']
        pesel = request.form['pesel']
        haslo = request.form['haslo']
        historia = request.form['historia']

        try:
            conn = sqlite3.connect('medica_plus.db')
            cursor = conn.cursor()
            cursor.execute('INSERT INTO pacjenci (imie, nazwisko, pesel, haslo, historia_medyczna) VALUES (?, ?, ?, ?, ?)',
                           (imie, nazwisko, pesel, haslo, historia))
            conn.commit()
            conn.close()
            return redirect(url_for('logowanie'))
        except sqlite3.IntegrityError:
            error = "BŁĄD: Ten PESEL jest już w bazie!"

    form_html = f"""
        <h2>Rejestracja Pacjenta</h2>
        <p class="error">{error}</p>
        <form method="post">
            <input name="imie" placeholder="Imię" required>
            <input name="nazwisko" placeholder="Nazwisko" required>
            <input name="pesel" placeholder="PESEL" required>
            <input name="haslo" type="password" placeholder="Hasło" required>
            <textarea name="historia" placeholder="Historia chorób i wyniki badań "></textarea>
            <button type="submit">Załóż konto</button>
        </form>
    """
    return render_template_string(LAYOUT, content=form_html)

@app.route('/logowanie', methods=['GET', 'POST'])
def logowanie():
    error = ""
    if request.method == 'POST':
        pesel = request.form['pesel']
        haslo = request.form['haslo']

        conn = sqlite3.connect('medica_plus.db')
        cursor = conn.cursor()
        cursor.execute('SELECT id FROM pacjenci WHERE pesel = ? AND haslo = ?', (pesel, haslo))
        user = cursor.fetchone()
        conn.close()

        if user:
            session['user_id'] = user[0]
            return redirect(url_for('karta'))
        error = "Błędny PESEL lub hasło."

    login_html = f"""
        <h2>Logowanie Pacjenta</h2>
        <p class="error">{error}</p>
        <form method="post">
            <input name="pesel" placeholder="PESEL" required>
            <input name="haslo" type="password" placeholder="Hasło" required>
            <button type="submit">Zaloguj się</button>
        </form>
    """
    return render_template_string(LAYOUT, content=login_html)

@app.route('/karta')
def karta():
    if 'user_id' not in session:
        return redirect(url_for('logowanie'))

    conn = sqlite3.connect('medica_plus.db')
    cursor = conn.cursor()
    cursor.execute('SELECT imie, nazwisko, pesel, historia_medyczna FROM pacjenci WHERE id = ?', (session['user_id'],))
    user = cursor.fetchone()
    conn.close()

    # Prezentacja danych zgodnie z wymogiem ochrony prywatności i RODO [cite: 27, 28]
    card_html = f"""
        <h2>Karta Pacjenta Medica+</h2>
        <p><strong>Imię i Nazwisko:</strong> {user[0]} {user[1]}</p>
        <p><strong>PESEL:</strong> {user[2]}</p>
        <hr>
        <h3>Historia Medyczna:</h3>
        <p style="background: #f0f0f0; padding: 10px;">{user[3] if user[3] else 'Brak wpisów.'}</p>
        <br>
        <a href="/logout"><button style="background: #666;">Wyloguj</button></a>
    """
    return render_template_string(LAYOUT, content=card_html)

@app.route('/logout')
def logout():
    session.clear()
    return redirect(url_for('index'))

# --- 4. URUCHOMIENIE W COLAB ---
serve_kernel_port_as_window(5000)
app.run(port=5000)
```
